```python
dc=dfv.Close-dfv.Open
#dh=dfv.High.shift(1)-dfv.High
f=plt.figure(num=None, figsize=(20, 5), dpi=80, facecolor='w', edgecolor='k')
h=pd.DataFrame()
h['dpH']=dph['dph']
h['H']=dfv.Close.shift(1)-dfv.Close
h['dc']=dc.shift(1)
h['n']=dc-dc
plt.plot(h[100:200])
```

```python
h
len(h)
```

```python
up,dn = [],[]
for i in range(0,len(h)):
    if h.loc[i]['dpH']>=0 and h.loc[i]['dc']>=0:
        up.append(h.loc[i]['dc'])
    if h.loc[i]['dpH']<0 and h.loc[i]['dc']<0:
        dn.append(h.loc[i]['dc'])
print ('long$',sum(up))
print ('short$',sum(dn))
print ('long success:',sum(up)/len(h))
print ('short success:',sum(dn)/len(h))
print ('right%',(len(up)+len(dn))/len(h))
print ('wrong:',len(h)-len(up)-len(dn))
print ('wrong%',(len(h)-len(up)-len(dn))/len(h))
```
